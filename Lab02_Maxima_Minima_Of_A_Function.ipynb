# Basic Trig Function Solve (With Explanation)

```python
#Importing the sympy library
import sympy as sp
from sympy import *
init_printing(pretty_print=True)
#Defining the mathematical function by using sp.symbols()
x,y,z=sp.symbols('x y z')
expr1=sp.sin(x)
#This is just to check out the graph of the mathematical function(This is not a must at all!)
sp.plot(expr1)
#Differentiating the given function
diff1=sp.diff(expr1,x)
#Equating the derivative function to 0 and obtaining the value at which minima and maxima occurs
check1=sp.solve(diff1,x)
#Differentiating the function one more to determine at what value of x do the maxima and minima occurs
diff2=sp.diff(diff1,x)
#Evaluating and storing exactly at what values of x do the maxima and minima occurs
maxmin1=diff2.subs(x,check1[0])
maxmin2=diff2.subs(x,check1[1])
#Storing the value at which maxima and minima occurs
if maxmin1<0:
  max=check1[0]
  min=check1[1]
elif maxmin1>0:
    max=check1[1]
    min=check1[0]
#Substituting the value at which maxima and minima occurs in the given function we are working with
maxval=expr1.subs(x,max)
minval=expr1.subs(x,min)
#this print(f"") is just a fancy formatting technique. You can print it as you please! 
print(f"Maxima is {maxval} and Minima is {minval}")
```

# Basic Polynomial Solve

```python
import sympy as sp
from sympy import *
init_printing(pretty_print=True)

x,y,z=sp.symbols('x y z')

expr1=x**3-5*x**2+3*x+2
sp.plot(expr1)

diff1=sp.diff(expr1,x)
check1=sp.solve(diff1,x)
print(check1)
diff2=sp.diff(diff1,x)
maxmin1=diff2.subs(x,check1[0])
maxmin2=diff2.subs(x,check1[1])

if maxmin1<0:
  max=check1[0]
  min=check1[1]
elif maxmin1>0:
    max=check1[1]
    min=check1[0]

maxval=expr1.subs(x,max)
minval=expr1.subs(x,min)
print(f"Maxima is {maxval} and Minima is {minval}")
```

#Overcoming Limitations of the Second Code

Remember, this is a fourth order polynomial hence, the first derivative of the function will be a third order polynomial. So the value at which maxima/minima will occur would have more than 2 values. It is imperative for you to iterate through all the values. That is why a for loop is introduced to iterate through all the values. 

```python
import sympy as sp
from sympy import *
init_printing(pretty_print=True)

x,y,z=sp.symbols('x y z')

expr1=x**4-8*x**3+22*x**2-24*x+5
sp.plot(expr1)

diff1=sp.diff(expr1,x)
check1=sp.solve(diff1,x)
print(check1)
diff2=sp.diff(diff1,x)
#The loop to iterate through all the values at which maxima and minima might occur
for k in range(0,len(check1)):
  maxmin=diff2.subs(x,check1[k])
  if maxmin<0:
    max=check1[k]
  elif maxmin>0:
    min=check1[k]  

maxval=expr1.subs(x,max)
minval=expr1.subs(x,min)
print(f"Maxima is {maxval} and Minima is {minval}")
```

#Overcoming Limitations of the Second Code- Part 2

If the solution(determining the root of the equation that is obtained upon first derivative) includes an imaginary number, than blindly calling up numbers from one after another would result in a traceback. Hence, it is imperative to check if the solution include any imaginary number. 

```python
import sympy as sp
from sympy import *
init_printing(pretty_print=True)

x,y,z=sp.symbols('x y z')

expr1=x**4-2/3*x**3-2*x**2+2*x

diff1=sp.diff(expr1,x)
diff2=sp.diff(diff1,x)
check1=sp.solve(diff1,x)
print(check1)
soln=[]
for i in range(0,len(check1)):
  if im(check1[0])!=0:
    continue
  else:
    soln.append(check1[i])

if len(soln)>0:
  for k in range(0, len(soln)):
    maxmin1=diff2.subs(x,soln[k])
    if maxmin1<0:
      max=soln[k]
      maxval=expr1.subs(x,max)
    elif maxmin1>0:
      min=soln[k]
      minval=expr1.subs(x,min)
  print(f"Maxima is {maxval} and Minima is {minval}")
else:
  print('No maxima or minima is found')
```
